```python
import json, os, sys
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import seaborn as sns


class NpEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.integer):
            return int(obj)
        if isinstance(obj, np.floating):
            return float(obj)
        if isinstance(obj, np.ndarray):
            return obj.tolist()
        return super(NpEncoder, self).default(obj)
    
def load_json(filepath):
    with open(filepath, "r", encoding="utf-8") as f:
        data = f.read()
        data = json.loads(data.strip())

    return data
def save_json (path, data):
    json_string = json.dumps(data, cls= NpEncoder)
    with open(path, 'w') as outfile:
        outfile.write(json_string)
```

```python
import matplotlib.pyplot as plt
import numpy as np

colors = ['r', 'b', 'y', 'k', 'g', 'c']
lines = ['-', '-', '-', '-', '-', '-']

    
def plot_dist_normlized (files, legends = None, label_tag = "confidence", title = "confidence"):
    hists = None 
    for i, file in enumerate (files):
        data = load_json (file)
        labels = ([d[label_tag] for d in data])
        y, binEdges = np.histogram(labels, bins=100)
        binEdges = binEdges/max (binEdges)
        y = y / max (y)
        bincenters = 0.5 * (binEdges[1:] + binEdges[:-1])
        plt.plot(bincenters, y, colors[i]+lines[i])
        
    plt.legend (legends)
    plt.title (title)
    plt.xlabel("Confidence score")
    plt.ylabel("Ratio of samples")
    result_path = "logs/images/"+ title +"_".join (legends) + ".eps"
    plt.savefig (result_path)
    
    plt.show()
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import RocCurveDisplay

colors = ['r', 'b', 'y', 'k', 'g', 'c']
lines = ['-', '-', '-', '-', '-', '-']


def plot_roc (files, legends = None, title = "roc"):
    
    fig, ax = plt.subplots(figsize=(6, 6))
    
    for i, (file1, file2) in enumerate (files):
        _id = load_json (file1)
        _id = [float(x['confidence']) for x in _id]
        pos = [1]*len(_id)

        _od = load_json (file2)
        _od = [float(x['confidence']) for x in _od]
        neg = [0]*len(_od)

        y = pos + neg
        confidences = _id + _od

        RocCurveDisplay.from_predictions(y, confidences, name=f"ROC curve "+ legends[i], color=colors[i], ax=ax)
        
    #plt.legend (legends)
    plt.title (title)
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    result_path = "logs/images/" + title +"_".join (legends) + "_roc.jpg"
    plt.savefig (result_path)
    
    plt.show()
    
```

```python
models = ["vgg16", "googlenet", "dit", "layoutlmv2", "layoutlmv3", "gvdoc"]
legends = [ "VGG-16", "GoogLeNet", "DiT", "LayoutLMv2", "LayoutLMv3", "GVdoc"]

files = [[f"logs/rvlcdip_{m}.json", f"logs/ood_{m}.json"] for m in models]
plot_roc(files, legends =legends, title = "ROC:T-O")
    
```

```python
import math 
from numpy import exp

def energy(vector, T = 1):
    e = -1 * T * math.log (exp(1* np.array(vector)/T).sum())
    return e 



def plot_energy (files, legends = None, label_tag = "logits", title = ""):
    hists = None 
    for i, file in enumerate (files):
        data = load_json (file)
        labels = ([energy(d[label_tag]) for d in data])
        y, binEdges = np.histogram(labels, bins=100)
        y = y / max (y)
        bincenters = 0.5 * (binEdges[1:] + binEdges[:-1])
        plt.plot(bincenters, y, colors[i]+lines[i])
        
    plt.legend (legends)
    plt.title (title)
    result_path = "logs/images/energy_"+ title +"_".join (legends) + "_dist.eps"
    print (result_path)
    plt.savefig (result_path)
    
    plt.show()
```

```python
models = ["vgg16", "googlenet", "dit", "layoutlmv2", "layoutlmv3", "gvdoc"]
titles = [ "VGG-16", "GoogLeNet", "DiT", "LayoutLMv2", "LayoutLMv3", "GVdoc"]

for m, title in zip (models, titles):
    file1 = f"logs/idood_{m}.json"
    file2 = f"logs/ood_{m}.json"
    file3 = f"logs/rvlcdip_{m}.json"
    
    plot_energy ([file1, file2], legends = ["RN", "RO"], title = f"{title}: RN vs RO")

    plot_energy ([file3, file2], legends = ["RT", "RO"], title = f"{title}: RT vs RO")
```
